```python
import pandas as pd
import numpy as np
```

```python
pat_17 = pd.read_excel('2017-2019_서울및6개광역시_아토피_최종_전처리완료.xlsx')
pat_20 = pd.read_excel('2020-2021_서울및6개광역시_아토피_최종_전처리완료.xlsx')
```

```python
hum_17 = pd.read_csv('humidity171819.csv')
hum_20 = pd.read_csv('humidity2021.csv')
```

```python
wind_17 = pd.read_csv('wind_17.csv', encoding='cp949')
wind_20 = pd.read_csv('wind_20.csv', encoding='cp949')
```

```python
temp_17 = pd.read_csv('Temperature_2017_2019.csv')
temp_20 = pd.read_csv('Temperature_2020_2021.csv')
```

```python
pol_17 = pd.read_csv('대기오염17-19.csv')
pol_20 = pd.read_csv('대기오염20-21.csv', encoding='cp949')
```

```python
df_17 = pd.merge(temp_17, hum_17, on=['날짜', '지역'])
df_20 = pd.merge(temp_20, hum_20, on=['날짜', '지역'])
```

```python
df_17 = pd.merge(df_17, wind_17, on=['날짜', '지역'])
df_20 = pd.merge(df_20, wind_20, on=['날짜', '지역'])
```

```python
df_17 = pd.merge(df_17, pol_17, on=['날짜', '지역'])
df_20 = pd.merge(df_20, pol_20, on=['날짜', '지역'])
```

```python
pat_17['날짜'] = pat_17['날짜'].astype(str)
pat_20['날짜'] = pat_20['날짜'].astype(str)
```

```python
df_17 = pd.merge(df_17, pat_17, on=['날짜', '지역'], how='outer')
df_20 = pd.merge(df_20, pat_20, on=['날짜', '지역'], how='outer')
```

```python
df_17['환자 수'] = df_17['환자 수'].fillna(0)
df_20['환자 수'] = df_20['환자 수'].fillna(0)
```

```python
df_17['년'] = df_17['날짜'].dt.year
df_20['년'] = df_20['날짜'].dt.year
```

## 양끝에 짤리는 데이터는 지우는게 좋을 것 같아서, 주가 1 또는 52인 데이터는 삭제하겠습니다.

```python
df_17.drop(df_17[df_17['주'].isin([1, 52])].index, inplace=True)
df_20.drop(df_20[df_20['주'].isin([1, 52])].index, inplace=True)
```

```python
df_17['풍향(N~NE)'] = (df_17['최대 풍속 풍향(16방위)'] > 0) & (df_17['최대 풍속 풍향(16방위)'] <= 45)
df_17['풍향(NE~E)'] = (df_17['최대 풍속 풍향(16방위)'] > 45) & (df_17['최대 풍속 풍향(16방위)'] <= 90)
df_17['풍향(E~SE)'] = (df_17['최대 풍속 풍향(16방위)'] > 90) & (df_17['최대 풍속 풍향(16방위)'] <= 135)
df_17['풍향(SE~S)'] = (df_17['최대 풍속 풍향(16방위)'] > 135) & (df_17['최대 풍속 풍향(16방위)'] <= 180)
df_17['풍향(S~SW)'] = (df_17['최대 풍속 풍향(16방위)'] > 180) & (df_17['최대 풍속 풍향(16방위)'] <= 225)
df_17['풍향(SW~W)'] = (df_17['최대 풍속 풍향(16방위)'] > 225) & (df_17['최대 풍속 풍향(16방위)'] <= 270)
df_17['풍향(W~NW)'] = (df_17['최대 풍속 풍향(16방위)'] > 270) & (df_17['최대 풍속 풍향(16방위)'] <= 315)
df_17['풍향(NW~N)'] = (df_17['최대 풍속 풍향(16방위)'] > 315) & (df_17['최대 풍속 풍향(16방위)'] <= 360)

df_20['풍향(N~NE)'] = (df_20['최대 풍속 풍향(16방위)'] > 0) & (df_20['최대 풍속 풍향(16방위)'] <= 45)
df_20['풍향(NE~E)'] = (df_20['최대 풍속 풍향(16방위)'] > 45) & (df_20['최대 풍속 풍향(16방위)'] <= 90)
df_20['풍향(E~SE)'] = (df_20['최대 풍속 풍향(16방위)'] > 90) & (df_20['최대 풍속 풍향(16방위)'] <= 135)
df_20['풍향(SE~S)'] = (df_20['최대 풍속 풍향(16방위)'] > 135) & (df_20['최대 풍속 풍향(16방위)'] <= 180)
df_20['풍향(S~SW)'] = (df_20['최대 풍속 풍향(16방위)'] > 180) & (df_20['최대 풍속 풍향(16방위)'] <= 225)
df_20['풍향(SW~W)'] = (df_20['최대 풍속 풍향(16방위)'] > 225) & (df_20['최대 풍속 풍향(16방위)'] <= 270)
df_20['풍향(W~NW)'] = (df_20['최대 풍속 풍향(16방위)'] > 270) & (df_20['최대 풍속 풍향(16방위)'] <= 315)
df_20['풍향(NW~N)'] = (df_20['최대 풍속 풍향(16방위)'] > 315) & (df_20['최대 풍속 풍향(16방위)'] <= 360)
```

```python
df_17.drop(['최대 풍속 풍향(16방위)', '풍정합(100m)'], axis=1, inplace=True)
df_20.drop(['최대 풍속 풍향(16방위)', '풍정합(100m)'], axis=1, inplace=True)
```

```python
df_17_tmp = df_17.groupby(['년', '주', '지역'])[['평균기온(°C)', '최저기온(°C)', '최고기온(°C)', '일교차(°C)', '평균 상대습도',
       '평균 풍속(m/s)', 'SO2_최대값', 'CO_최대값', 'O3_최대값', 'NO2_최대값',
       'PM10_최대값', 'PM25_최대값']].mean().reset_index()

df_20_tmp = df_20.groupby(['년', '주', '지역'])[['평균기온(°C)', '최저기온(°C)', '최고기온(°C)', '일교차(°C)', '평균 상대습도',
       '평균 풍속(m/s)', 'SO2_최대값', 'CO_최대값', 'O3_최대값', 'NO2_최대값',
       'PM10_최대값', 'PM25_최대값']].mean().reset_index()
```

```python
df_17_temp = pd.merge(df_17_tmp, df_17_tmp2, on=['년', '주', '지역'])
df_17_final = pd.merge(df_17_temp, df_17_tmp3, on=['년', '주', '지역'])

df_20_temp = pd.merge(df_20_tmp, df_20_tmp2, on=['년', '주', '지역'])
df_20_final = pd.merge(df_20_temp, df_20_tmp3, on=['년', '주', '지역'])
```
